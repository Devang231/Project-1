```python
import matplotlib.pyplot as plt
import pandas as pd
import csv as csv



# Read data from CSV file
data = pd.read_csv('/Users/devangpatel/Desktop/Average Generation per Solar Farm.csv')

# Extract the categories and values from the data
categories = data['County'].tolist()
values = data['Average Generation per Solar Farm'].tolist()

colors = ['red', 'green', 'blue', 'orange', 'purple', 'yellow']

# Create bar chart with different colors for each bar
plt.bar(categories, values, color=colors)

# Set chart title and axis labels
plt.title('Average Generation Per Solar Farm by County')
plt.xlabel('County')
plt.ylabel('Average Generation Per Solar Farm')

plt.xticks(rotation=45, ha='right')

plt.savefig('Average_Generation_Per_Solar_Farm_By_County_bar_chart.png')

# Show the chart
plt.show()
```
