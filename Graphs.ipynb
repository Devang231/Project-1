```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from scipy.stats import linregress


energyData = pd.read_csv("/Users/tanneramman/Downloads/energyDataLinear2021.csv")
energyData.head()
```

```python
energyData.columns
```

```python
energy = energyData.copy()
energy = energy.drop(columns = 'Unnamed: 0')
energy.head()
```

```python
energy.columns
```

```python
countyGroup = energy.groupby(['County'], as_index = False).mean()

justEnergy = countyGroup[['County', 'Estimate!!Total!!Utility gas', 
                           'Estimate!!Total!!Bottled, tank, or LP gas', 'Estimate!!Total!!Electricity',
                           'Estimate!!Total!!Fuel oil, kerosene, etc.', 'Estimate!!Total!!Coal or coke',
                           'Estimate!!Total!!Wood', 'Estimate!!Total!!Solar energy', 
                           'Estimate!!Total!!Other fuel', 'Estimate!!Total!!No fuel used']]
copy = justEnergy.copy()
copy['summed'] = copy.sum(axis=1)


x = countyGroup['County']
y0 = countyGroup['Estimate!!Total!!Bottled, tank, or LP gas']/copy['summed']*100
y1 = countyGroup['Estimate!!Total!!Utility gas']/copy['summed']*100
y2 = countyGroup['Estimate!!Total!!Bottled, tank, or LP gas']/copy['summed']*100
y3 = countyGroup['Estimate!!Total!!Electricity']/copy['summed']*100
y4 = countyGroup['Estimate!!Total!!Fuel oil, kerosene, etc.']/copy['summed']*100
y5 = countyGroup['Estimate!!Total!!Coal or coke']/copy['summed']*100
y6 = countyGroup['Estimate!!Total!!Wood']/copy['summed']*100
y7 = countyGroup['Estimate!!Total!!Solar energy']/copy['summed']*100
y8 = countyGroup['Estimate!!Total!!Other fuel']/copy['summed']*100
y9 = countyGroup['Estimate!!Total!!No fuel used']/copy['summed']*100                       

plt.bar(x, y0, color='r')
plt.bar(x, y1, bottom=y0, color='b')
plt.bar(x, y2, bottom=y0+y1, color='y')
plt.bar(x, y3, bottom=y0+y1+y2, color='g')
plt.bar(x, y4, bottom=y0+y1+y2+y3, color='r')
plt.bar(x, y5, bottom=y0+y1+y2+y3+y4, color='c')
plt.bar(x, y6, bottom=y0+y1+y2+y3+y4+y5, color='m')
plt.bar(x, y7, bottom=y0+y1+y2+y3+y4+y5+y6, color='k')
plt.bar(x, y8, bottom=y0+y1+y2+y3+y4+y5+y6+y7, color='y')
plt.bar(x, y8, bottom=y0+y1+y2+y3+y4+y5+y6+y7+y8, color='b')
plt.xticks(rotation = 45)
plt.savefig("/Users/tanneramman/Downloads/EnergyBarStacked.png")
#plt.xlabel("Teams")
#plt.ylabel("Score")
#plt.legend(["Round 1", "Round 2", "Round 3", "Round 4"])
#plt.title("Scores by Teams in 4 Rounds")
plt.show()
```
