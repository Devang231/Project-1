```python
import geopandas as gpd
import matplotlib.pyplot as plt
```

```python
import pandas as pd
```

```python
counties = gpd.read_file('/Users/tanneramman/Downloads/Colorado_County_Boundaries/Colorado_County_Boundaries.shp')
pop_data = pd.read_csv('/Users/tanneramman/Downloads/PopulationByCounty.csv')
pop_data = pop_data.drop(columns = 'Unnamed: 2')
pop_data = pop_data.rename(columns = {'County' : 'LABEL'})
pop_data.head()
```

```python
counties = counties.merge(pop_data, on='LABEL')
```

```python
counties
```

```python
fig, ax = plt.subplots(figsize=(10,10))
counties.plot(column='Population', cmap='viridis', legend=True, ax=ax)

from matplotlib.lines import Line2D
legend_elements = [
    Line2D([0], [0], marker='o', color='w', label='< 100,000', markerfacecolor='indigo', markersize=10),
    Line2D([0], [0], marker='o', color='w', label='100,000 - 500,000', markerfacecolor='y', markersize=10),
    Line2D([0], [0], marker='o', color='w', label='> 500,000', markerfacecolor='g', markersize=10)
]
ax.legend(handles=legend_elements, loc='lower left', title='Population Categories')
plt.title("County Populations")

plt.savefig("/Users/tanneramman/Downloads/PopulationChloreo.pdf", bbox_inches='tight')
```
