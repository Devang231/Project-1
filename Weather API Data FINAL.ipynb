```python
#Reading in counties DF
counties = pd.read_csv('C:/Users/adina/Desktop/Classwork/Project 1/Project-1/CO County Data.csv',encoding='cp1252')
counties_df = pd.DataFrame(counties)
counties_df.head()
```

```python
# Create list of dates (Mondays in a year)
# Following code adapted from: https://stackoverflow.com/questions/2003870/how-can-i-select-all-of-the-sundays-for-a-year-using-python

def allmondays(year):
    return pd.date_range(start=str(year), end=str(year+1), 
                         freq='W-MON').strftime('%Y-%m-%d').tolist()

dates = allmondays(2020)
dates
```

```python
#Dates ranges for batches
for date in dates[39:52]:
    print(date)
    
# 0-13
# 13-26
# 26-39
# 39-52
```

```python
#Gather the data in chunks
#Update the date range and the api key to limit the number of calls per key

weather_data = []

# loop through dates in chunks
for date in dates[39:52]:
    
    # for index, row in counties_weather_df.iterrows():
    for index, row in counties_df.iterrows():

    # Get latitude, longitude from the DataFrame
        latitude = row['Latitude']
        longitude = row['Longitude']
        request_date = date
        api_key = tanner_api_key
        county = row['County']

            # Set base URL
        base_url = f'https://weather.visualcrossing.com/VisualCrossingWebServices/rest/services/timeline/{latitude},{longitude}/{request_date}/{request_date}?key={api_key}&contentType=json&include=days&elements=cloudcover,datetime,solarradiation,solarenergy,tempmax,tempmin,temp,humidity,precipcover,preciptype,uvindex,sunrise,sunset'

            # Make an API request and convert to JSON
        response = requests.get(base_url).json()

        try:
            weather_dict = response['days'][0]
            weather_dict['County'] = county
            weather_data.append(weather_dict)
        except:
            pass

weather_data
```

```python
#Convert to DF and confirm correct shape
part4_df = pd.DataFrame(weather_data)
part4_df.shape
```
