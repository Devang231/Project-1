```python
import pandas as pd
```

```python
#Read in all CSVs

county_data = pd.read_csv("C:/Users/adina/Desktop/Classwork/Project 1/Project-1/CO County Data.csv",encoding = "ISO-8859-1")
weather_data = pd.read_csv("C:/Users/adina/Desktop/Classwork/Project 1/Project-1/Weather API Data CSVs/20_21_22_Mondays.csv")
```

```python
#Display county data

county_data.head()
```

```python
# Calculate averages for weather data by county and make new data frame

weather_data_avg = weather_data.groupby('County').mean().reset_index()
weather_data_avg.head()
```

```python
# Combine census data into one df

census_energy_2017 = pd.read_csv("C:/Users/adina/Desktop/Classwork/Project 1/Project-1/Census Energy Data CSV/2017census_data.csv")
census_energy_2018 = pd.read_csv("C:/Users/adina/Desktop/Classwork/Project 1/Project-1/Census Energy Data CSV/2018census_data.csv")
census_energy_2019 = pd.read_csv("C:/Users/adina/Desktop/Classwork/Project 1/Project-1/Census Energy Data CSV/2019census_data.csv")
census_energy_2020 = pd.read_csv("C:/Users/adina/Desktop/Classwork/Project 1/Project-1/Census Energy Data CSV/2020census_data.csv")
census_pop_2017 = pd.read_csv("C:/Users/adina/Desktop/Classwork/Project 1/Project-1/Census Population Data/2017Pop_census_data.csv")
census_pop_2018 = pd.read_csv("C:/Users/adina/Desktop/Classwork/Project 1/Project-1/Census Population Data/2018Pop_census_data.csv")
census_pop_2019 = pd.read_csv("C:/Users/adina/Desktop/Classwork/Project 1/Project-1/Census Population Data/2019Pop_census_data.csv")
census_pop_2020 = pd.read_csv("C:/Users/adina/Desktop/Classwork/Project 1/Project-1/Census Population Data/2020Pop_census_data.csv")

census_2017 = pd.merge(census_energy_2017,census_pop_2017,on='Name')
census_2018 = pd.merge(census_energy_2018,census_pop_2018,on='Name')
census_2019 = pd.merge(census_energy_2019,census_pop_2019,on='Name')
census_2020 = pd.merge(census_energy_2020,census_pop_2020,on='Name')

census_data = pd.concat([census_2017,census_2018,census_2019,census_2020])
```

```python
# Replace county names in census data to make them consistent with the other data frames and rename some columns

new_counties = []
for index, row in census_data.iterrows():
    county = row['Name']
    new_county = county.split(', ')[0].replace(' County','')
    new_counties.append(new_county)    

census_data['Name'] = new_counties
census_data = census_data.rename(columns={'County':'County Number'})
census_data = census_data.rename(columns={'Name':'County'})
census_data.head()
```

```python
# Merge all data on county name

all_data = pd.merge(census_data,county_data,on='County')
all_data = pd.merge(all_data,weather_data_avg,on='County')
all_data.head()
```

```python
# Export to CSV

all_data.to_csv("C:/Users/adina/Desktop/Classwork/Project 1/Project-1/CensusWeatherCounties.csv",index=False)
```
