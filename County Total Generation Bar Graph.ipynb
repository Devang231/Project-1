```python
import matplotlib.pyplot as plt
import pandas as pd
import csv as csv



# Read data from CSV file
data = pd.read_csv('/Users/devangpatel/Desktop/Generation Total by County.csv')

# Extract the categories and values from the data
categories = data['County'].tolist()
values = data['Total_Generation'].tolist()

colors = ['red', 'green', 'blue', 'orange', 'purple', 'yellow']

# Create bar chart with different colors for each bar
plt.bar(categories, values, color=colors)

# Set chart title and axis labels
plt.title('Total Generation by County')
plt.xlabel('County')
plt.ylabel('Total Generation')

plt.xticks(rotation=45, ha='right')

plt.savefig('County_Generation_bar_chart.png')

# Show the chart
plt.show()


```
